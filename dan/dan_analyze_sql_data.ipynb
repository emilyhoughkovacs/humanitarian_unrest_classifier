```python
def makeSQLQuery2(table_names):
    '''list of table names -- > SQL query to pull a table for which the columns are:
    
        countrycode, year, [table_names]
        
        Joins thsoe tables on countrycode and year, one at a time
        
        '''
    current_unit = 'a'
    
    query = '(SELECT ' + table_names[0] + '.countrycode AS countrycode, ' + \
            table_names[0] + '.year AS year, ' + \
            table_names[0] + '.value AS ' + table_names[0] + ', '  +\
            table_names[1] + '.value AS ' + table_names[1] + \
            ' FROM ' + table_names[0] + \
            ' JOIN ' + table_names[1] + \
            ' ON ' + table_names[0] + '.countrycode=' + table_names[1] + '.countrycode' + \
            ' AND ' + table_names[0] + '.year=' + table_names[1] + '.year)'

    
    for name in table_names[2:]:
        stem = '(SELECT a.*, ' + name + '.value AS ' + name + ' FROM'
        tail = ' AS a JOIN ' + name + \
                ' ON a.countrycode = ' + name + '.countrycode ' \
                'AND a.year = ' + name + '.year)'
        
        query = stem + query + tail
    
    #if joining with disasters, add this code
    stem2 = '(SELECT a.*, disaster, homeless, countryname FROM '
    tail2 = ' AS a JOIN disasters ON a.year = disasters.year AND a.countrycode = disasters.countrycode)'
    query = stem2 + query + tail2
    
    return query
```

```python
df['binary'] = df['homeless'].apply(lambda x: 1 if x > 0 else 0)
```

```python
col_names = ['adolescent_fertility', \
               'cell_phones_per100', \
               'elec_per_capita', \
               'gdp_growth', \
               'have_good_sanitation', \
               'internet_per100', \
               'labor_force_participation', \
#                'population_total', \
               'rural_pop_in_poverty', \
               'unemployment', \
#                'vulnerable_employment_f', \
               'vulnerable_employment_m'
            ]
```

```python
#X values should be all the columns with our demographic variables (see table_names)
X = df[col_names]

#y is a binary representation of whether or not people were displaced by a natural disaster
y = df['binary']
```

```python
from sklearn.cross_validation import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(X, y, test_size=0.30, random_state=4444)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, average_precision_score, f1_score
```

```python
names, accs, prec, rec, f1s = [], [], [], [], []
for algorithm in (LogisticRegression, KNeighborsClassifier, \
                  GaussianNB,SVC, DecisionTreeClassifier, RandomForestClassifier):
    model = algorithm()
    model.fit(xtrain, ytrain)
    
#     print(algorithm.__name__, '\n', classification_report(ytest, model.predict(xtest), '\n')
          
    ypredict = model.predict(xtest)

    names.append(algorithm.__name__)
    accs.append(accuracy_score(ytest, model.predict(xtest)))
    prec.append(precision_score(ytest, model.predict(xtest)))
    rec.append(recall_score(ytest, model.predict(xtest)))
    f1s.append(f1_score(ytest, model.predict(xtest)))
```

```python
pd.DataFrame({' Algorithm' : names, \
             'Accuracy Score': accs, \
             'Precision Score': prec, \
              'Recall score' : prec, \
             'F1 score': f1s})
```
